```python
import csv, sqlite3
import pandas as pd
```

```python
con = sqlite3.connect("med_cabinet_index.db")
cur = con.cursor()
cur.execute("CREATE TABLE strains (i, Strain, Type, Rating, Effects, Flavor, Description, full_descriptions);") # use your column names here
```

```python
# cur.execute("DROP TABLE strains")
```

```python
df = pd.read_csv('full_cannabis_descriptions_index.csv')
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
for index, row in df.iterrows():
    command = """INSERT INTO strains (i, Strain, Type, Rating, Effects, Flavor, Description, full_descriptions)
                VALUES """
    data = str((df['index'][index], df['Strain'][index], df['Type'][index], df['Rating'][index], df['Effects'][index],
               df['Flavor'][index], df['Description'][index], df['full_descriptions'][index]))
    command = command + data
    command = command + ';'
    
    # Execute command
    cur.execute(command)
```

```python
con.commit()
con.close()
```
