```python
from fuzzywuzzy import process

def find_strain_named(df, df2):
    df = df.copy()
    df2 = df2.copy()
    
    labs = []
    
    for strain in df['Strain']:
        str2Match = strain
        strOptions = df2['strain'].to_list()
        highest = process.extract(str2Match,strOptions)
        labs.append(highest)
    
    return labs

labs = find_strain_named(pot, results_all)
```

```python
str2Match = pot['Strain'][0]
strOptions = results_all['strain'].to_list()
highest = process.extract(str2Match,strOptions)
print(highest)
```

```python
labs_filtered = []

for lab in labs:
    if lab[0][1] >= 95:
        labs_filtered.append(lab[0][0])
    else:
        labs_filtered.append(float('nan'))
```
