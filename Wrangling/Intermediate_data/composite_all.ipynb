```python
from fuzzywuzzy import process

def find_strain_named(df, df2):
    df = df.copy()
    df2 = df2.copy()
    
    labs = []
    
    for strain in df['Strain']:
        str2Match = strain
        strOptions = df2['strain'].to_list()
        highest = process.extract(str2Match,strOptions)
        labs.append(highest)
    
    return labs

labs = find_strain_named(pot, results_all)
```

```python
labs
```

```python
labs_filtered = []

for lab in labs:
    if lab[0][1] >= 95:
        labs_filtered.append(lab[0][0])
    else:
        labs_filtered.append(float('nan'))
```

```python
len(labs_filtered)
```

```python
results_all[results_all['labs'] == lab]['strain']
```

```python
filtered_labs = []
for lab in labs_filtered:
    if type(lab) == str:
        filtered_labs.append(results_all[results_all['strain'] == lab]['strain'])
    else:
        pass
```

```python
pot['labs'] = labs_filtered
correct_strain_name = []
for lab in labs_filtered:
    if type(lab) == str:
        correct_strain_name.append(pot[pot['labs'] == lab]['Strain'])
    else:
        pass
```

```python
df = pd.DataFrame(columns=['index', 'Strain'])
index = []
for strain in filtered_labs:
    if type(strain) == float:
        pass
    else:
        index.append(strain.index[0])
```

```python
strains = []
for strain in correct_strain_name:
    if type(strain) == float:
        pass
    else:
        strains.append(strain.iloc[0])
```

```python
df['index'] = index
df['Strain'] = strains
```

```python
more_data = pd.merge(df, results_all, left_on='index', right_on=results_all.index, how='left')
```
